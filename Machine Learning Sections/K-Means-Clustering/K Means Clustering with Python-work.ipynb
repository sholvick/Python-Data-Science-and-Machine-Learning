```python
# K Means Clustering is UN-supervised learning

import seaborn as sns
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# import sklearn tool for generating artificial data

from sklearn.datasets import make_blobs
```

```python
# generate artificial data

data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
# explore the data

data[0].shape
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
data[1]  # these are the labels
```

```python
# use sklearn to create a K Means cluster algorithm

from sklearn.cluster import KMeans
```

```python
# instantiate the model, indicating # of clusters
kmeans = KMeans(n_clusters=4)

# fit the model to the features
kmeans.fit(data[0])

# show the centers
kmeans.cluster_centers_
```

```python
kmeans.labels_   # these are the predicted labels, compare to data[1] above
```

```python
# explore the number of clusters by plotting - show how predicted labels (kmeans.labels_) compares to the generated labels (data[1])

fig , (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))    # sharey=True: both plots share the same y axis

ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
