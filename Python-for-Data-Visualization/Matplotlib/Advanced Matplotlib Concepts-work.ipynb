```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
# If not using jupyter notebook, use:  plt.show() to display the visualization
```

```python
# simple example using 2 numpy arrays

x = np.linspace(0,5,11)
y = x **2
```

```python
x
```

```python
y
```

```python
# functional method to create a plot

plt.plot(x,y)
plt.show()  # not needed here
```

```python
# add matlab arguments, like linestiles and colors
plt.plot(x,y,'r-')
```

```python
# add x/y labels and title
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
# create multiple plots on the same canvas

# create a subplot
plt.subplot(1,2,1)  # (num_rows, num_cols, plot_num_referring_to)
plt.plot(x,y,'r')

# add another subplot
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# Object Oriented Method to create plots

# Create figure objects, then call methods on them
fig = plt.figure()  # blank canvas

# add axis
axes = fig.add_axes([0.1,0.1,0.8,0.8])  # left, bottome, width, height

# plot on the new axes
axes.plot(x,y)

# add x/y labes and title
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Set Title')
```

```python
# now create 2 visualizations on 1 canvas

# create the empty figure (canvas)
fig = plt.figure()

# Create 2 blank axes / chart on the figure
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

# add plots into the blank axes (plots) we just created, with titles
axes1.plot(x,y)
axes1.set_title('LARGER PLOT')

axes2.plot(y,x)
axes2.set_title('SMALLER PLOT')
```

```python
# use plt.subplots() oo method

fig,axes = plt.subplots(nrows=3, ncols=3)   # note this is 'tuple unpacking'
# eliminate overlaps
plt.tight_layout()
#axes.plot(x,y)
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)
axes  # note that the axes object is just an array of axes objects, so we can iterate through it...
```

```python
# iterate through axes objects
fig,axes = plt.subplots(nrows=1,ncols=2)

for current_ax in axes:
    current_ax.plot(x,y)
```

```python
# we can also use indexing to identify plot
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[1].plot(y,x)

axes[0].set_title('First Plot')
axes[1].set_title('Second Plot')
```

```python
fig = plt.figure(figsize=(8,2))  # figsize(width_inch,height_inch)

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
# do the same thing with subplots
fig,axes = plt.subplots(nrows=2,ncols=1, figsize=(8,2))  # figsize(width_inch,height_inch)

axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# use legends to use label text to clarify which plot is which

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2, label = 'X Squared')
ax.plot(x,x**3, label = 'X Cubed')

# add a legend, 2-step process  - requires each plot to have a label defined
ax.legend(loc='best')  # loc is optional
```

```python
# Set colors
# create a figure
fig = plt.figure()

# add axes
ax = fig.add_axes([0,0,1,1])

# set color of line, width
ax.plot(x,y,color='orange', linewidth=5, alpha=0.5, linestyle='-.', marker=5, markersize=20, 
        markerfacecolor='purple', markeredgewidth=3, markeredgecolor='green') 

    # color = color name, or RGB value like #FF8C00
    # linewidth = width of line, 1 is default, 5 = 5 * standard width
    # alpha = transparency, set as float 0.5
    # linestyle, or ls = -,--,-.,step
    # marker = x,o,+,*, numbercodes 1,2,3,4,5,etc
    # markersize
    # markerfacecolor - color inside the marker (not counting outline)
    # markeredgewidth
    # markeredgecolor
```

```python
# adjust upper and lower bounds of the plot

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color='purple',lw=2,ls='--')

# set bounds
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
