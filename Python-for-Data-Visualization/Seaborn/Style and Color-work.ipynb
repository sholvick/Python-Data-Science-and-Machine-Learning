```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

tips = sns.load_dataset('tips')
tips.head()

### Basic Overview:
# use palette to change color palette
# use set_context to change font size using font_scale
# use figsize and figure to change size of your figure
```

```python
# create a count plot
sns.countplot(x='sex',data=tips)
```

```python
# define style

sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('darkgrid')
sns.countplot(x='sex',data=tips)
```

```python
# remove spine using despine

sns.set_style('ticks')
sns.countplot(x='sex',data=tips)

sns.despine()
```

```python
# specify to remove left and bottom spines

sns.set_style('ticks')
sns.countplot(x='sex',data=tips)

sns.despine(left=True,bottom=True)
```

```python
# size and aspect using matplotlib figsize

plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
# scale and context using set_context - to override default font scale

sns.set_context('poster',font_scale=3)  # font_scale is a multiplier: 3 = 3 x normal font size
sns.countplot(x='sex',data=tips)
```

```python
# specify palette or color

# first, reset font and scale
sns.set_context('paper',font_scale=1)

sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='seismic')

# palette options can be found searching google for matplotlib colormaps
url = 'https://matplotlib.org/examples/color/colormaps_reference.html'
```
