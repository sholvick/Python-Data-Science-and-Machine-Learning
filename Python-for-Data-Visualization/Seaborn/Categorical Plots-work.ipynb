```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# Categorical plots for seeing distributions of categorical columns (sex, smoker, day, time, etc.) in reference 
# to either another categorical column or a numeric column
```

```python
# bar plot - general plot that allows you to aggregate the categorical data off some function (default is mean)
# think of it as a visualization of a group-by action

sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
# add or change the estimator object if you want affect the aggregating or grouping - default='mean'
import numpy as np

sns.barplot(x='sex',y='total_bill',data=tips, estimator=np.std)
# estimator = np.std = use numpy standard deviation
```

```python
# count plot - same as a bar plot, except estimator is explicitly counting the number of occurrences, so you only give x value
# so y axis is always going to be 'count'

sns.countplot(x='sex',data=tips)
```

```python
# box and violin plots - used to show the distribution of categorical plot

# box plot (aka box and whisker plot) -  shows distribution of quantitative data that hopefully facilitates 
# comparison between variables

sns.boxplot(x='day',y='total_bill',data=tips)  # x=categorical, y=numeric

# points outside the whiskers are determined to be 'outliers'
```

```python
# add hue to show more comparisons

sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker') 
```

```python
# violin plot - similar to box plot, tells same story
# shows distribution across some category

sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
# can also customize with more comparison using hue

sns.violinplot(x='day',y='total_bill',data=tips, hue='sex')
```

```python
# or split each violin object in half based on the hue to see true side-by-side comparison

sns.violinplot(x='day',y='total_bill',data=tips, hue='sex', split=True)
```

```python

###  Advanced methods / plots
```

```python
# strip plot - draw a scatter plot where one variable is categorical

sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
# can be difficult to see how many dots are stacked - use 'jitter' option to add some vertical separation of points to see volume

sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
# can also add hue and split to further segment
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True, hue='sex', split=True)
```

```python
# swarm plot - basically combine a stripplot and a violin plot together
# points are adjusted automatically so they do not overlap

sns.swarmplot(x='day',y='total_bill',data=tips)

# not recommended, can be costly processor-wise, and can be hard to read if points are spread too far apart
```

```python
# can also plot a violin & swarmplot together

sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips, color='black')
```

```python
# factor plot - most general form of all these plots
# kind= explains the factor plot

sns.factorplot(x='day',y='total_bill', data=tips, kind='bar')
```

```python
# recreate the violin plot using factor plot
sns.factorplot(x='day',y='total_bill', data=tips, kind='violin')
```
