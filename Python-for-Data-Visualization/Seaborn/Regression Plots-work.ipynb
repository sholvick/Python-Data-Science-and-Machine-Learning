```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

```python
# use lmplot to create a very simple linear model plot

sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex', markers=['o','v'], scatter_kws={'s':100}) 
# since using hue to make 2 segments, can pass in a list of markers for each type
# scatter_kws sends parameters to underlying matplotlib to change attributes - 's':100 means make size 100, or larger
```

```python
# you can use a grid instead of separating by hue

sns.lmplot(x='total_bill',y='tip',data=tips, col='sex', row='time')

# col = separate by columns
# row = separate by rows
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips, col='day', hue='sex', aspect=0.6, size=8)
```
