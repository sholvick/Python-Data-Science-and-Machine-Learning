```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
# make an autoatic pair plot

sns.pairplot(iris)
```

```python
# customize with a lot more control over the plot

# call pair grid, which is sort of an empty pair plot

sns.PairGrid(iris)
```

```python
# add options to the pair grid

# specify plot types in each of the grid cells

g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
# facet grid

# use tips dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
# create an empty Facet Grid
g = sns.FacetGrid(data=tips,col='time',row='smoker')

# now define plot types
g.map(sns.distplot,'total_bill')
```

```python
# try as a scatter plot
g = sns.FacetGrid(data=tips,col='time',row='smoker')

# now define plot types
g.map(plt.scatter,'total_bill','tip')  # note: scatter requires an additional variable for comparison
```
