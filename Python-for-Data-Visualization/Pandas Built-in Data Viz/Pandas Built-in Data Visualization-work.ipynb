```python
import numpy as np
import pandas as pd
import seaborn as sns

%matplotlib inline

df1 = pd.read_csv('df1',index_col=0)

# time series dataframe
df1.head()
```

```python
df2 = pd.read_csv('df2')

# no index, just random numbers
df2.head()
```

```python
# make a histogram for all values on df1 col A
# note - charts will automatically look better if you just have Seaborn imported - it will inherit some nicer default formatting

df1['A'].hist(bins=30)
```

```python
# almost all plots can be created as a method on the dataframe itself by specifying 'kind'

# create histogram
df1['A'].plot(kind='hist',bins=30)
```

```python
# create a default histogram
df1['A'].plot(kind='hist')
```

```python
df2.head()
```

```python
# create a simple area plot
df2.plot.area()
```

```python
df2.plot.area(alpha=0.4)
```

```python
# create a bar plot

df2.plot.bar()
```

```python
# make it stacked

df2.plot.bar(stacked=True)
```

```python
# basic histogram again

df1['A'].plot.hist(bins=50)
```

```python
# line plot - must specify x,y

df1.plot.line(x=df1.index,y='B')
```

```python
# you can also add matplotlib arguments, since this is built off matplotlib

df1.plot.line(x=df1.index,y='B',figsize=(12,3),lw=1)
```

```python
# scatter plot - you can specify color (c) like hue

df1.plot.scatter(x='A', y='B', c='C', cmap='coolwarm')
```

```python
# you can also specify scatter plot points by size (s)

df1.plot.scatter(x='A', y='B', s=df1['C']*100)  # note: may have to multipy s (size) as it will be very small
```

```python
# box plots

df2.plot.box()
```

```python
# hexagonal bin plot for bivariate data (2 variables)

# make a new df for it to show up correctly
df = pd.DataFrame(np.random.randn(1000,2),columns=['a','b'])

df.head()
```

```python
# create hex plot

df.plot.hexbin(x='a', y='b', gridsize=25, cmap='coolwarm')
```

```python
# kernel density estimation (kde) plot

df2['a'].plot.kde()
```

```python
# density plot, similar to kde

df2['a'].plot.density()
```

```python
# you can also plot on all items in df, not just 'a'

df2.plot.density()
```
